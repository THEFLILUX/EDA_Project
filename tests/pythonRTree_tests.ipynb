```
import csv
from rtree import index

class Trip:
    def __init__(self, id):
        self.id = id
        self.Pickup_longitude = float(0)
        self.Pickup_latitude  = float(0)

filename = '/content/drive/My Drive/EDAFinalProject/green_tripdata_2015-01.csv'

index.Property().dimension = 2
index.Property().buffering_capacity = 110
rTree = index.Index("/content/drive/My Drive/EDAFinalProject/rtree_trip11", properties = index.Property())

with open(filename) as csv_file:
  csv_list = csv.reader(csv_file, delimiter=',')
  counter = 1
  next(csv_list)
  for register in csv_list:
        coord = []
        trip_obj = Trip(counter)
        coord.append(float(register[5]))
        coord.append(float(register[6]))
        coord.append(float(register[5]))
        coord.append(float(register[6]))
        rTree.insert(counter, coord, trip_obj)
        counter = counter + 1
rTree.close()
print(str(counter) + " trips")
```

```
from rtree import index

def query(p1, p2):
  rtree = "/content/drive/My Drive/EDAFinalProject/rtree_trip11"
  
  index.Property().dimension = 2
  index.Property().buffering_capacity = 110

  index_rtree = index.Rtree(rtree, properties=index.Property()) 

  result = [n.object for n in index_rtree.intersection((p1[0], p1[1], p2[0], p2[1]), objects=True)] 

  return len(result)

#-73.91, 40.80 | -73.96003, 40.814 | -73.9185, 40.70
#-73.90, 40.81 | -73.960025, 40.815 | -73.90, 40.80
print(str(query([-73.96003, 40.814], [-73.960025, 40.815])))
```
